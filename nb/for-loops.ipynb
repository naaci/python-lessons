# `for` Statements

```python
fib = 0,1,1,2,3,5,8,13,21
```

```python
for i in fib:
  print(i)
```

Like `if` and `while` statements, `for` statements can also have an optional `else` statement.

```python
for i in fib:
  print(i)
else:
  print("finished")
```

## Breaking The Loops
Sometimes you may want to exit from the loop even the condition was `True`.
To do use `break`.
If you break the loop then `else` statement is not evaluated.

```python
for i in fib:
  if i > 10:
    break
  print(i)
else:
  print("finished")
```

```python
for i in fib:
  if i > 10:
    continue
  print(i)
else:
  print("finished")
```

## Nested Loops
When using `break` inside a nested loop, you can break only the inner one.

```python
for i in fib:
  for i in fib:
    break
  print("Only the inner loop is break.")
  break
```

## Iterables
In Python strings, tuples, lists, sets, dictioanries are iterables.

```python
for a in 'hello world!':
  print(a,ord(a))
```

```python
letters_used = set()
for letter in 'hello world':
  letters_used.add(letter)
print(letters_used)
```

```python
letters_used = dict()
for letter in 'hello world':
  if letter in letters_used:
    letters_used[letter] += 1
    # letters_used[letter] = letters_used[letter] + 1
  else:
    letters_used[letter] = 1
print(letters_used)
```

### Iterating Over Dictioanies

```python
for letter in letters_used:
  print(letter,letters_used[letter],sep='\t')
```

```python
letters_used.items()
```

```python
for letter,count in letters_used.items():
  print(letter,count,sep='\t')
```

```python
for letter_count in letters_used.items():
  print(*letter_count,sep='\t')
```
