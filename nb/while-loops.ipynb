# `while` Statements
In Python `while` statement evaluates a condition then runs a block of code like `if` statement.
But unlike `if` it repeats that block of code **while** the condition is equals to `True`.

```python
name = input("enter your name: ")
while name == "":
  name = input("please enter your name: ")
```

Like `if` statements, `while` statements can also have an optional `else` statement.

```python
name = input("enter your name: ")
while name == "":
  name = input("please enter your name: ")
else:
  print("hello", name)
```

## Breaking The Loops
Sometimes you may want to exit from the loop even the condition was `True`.
To do use `break`.
If you break the loop then `else` statement is not evaluated.

```python
counter = 3
name = input("enter your name: ")
while name == "":
  counter = counter - 1
  if counter == 0:
    break
  name = input(f"please enter your name: (You have {counter} tries)")
else:
  print("hello", name)
```

```python
while not (name := input("enter your name: ")):
  pass
else:
  print("hello", name)
```

`continue` statement is used to go to the start of the loop without evaluating the remaining of the loop.

```python
a = 12
while a < 100:
  a = a + 3
  if a % 2 == 0:
    continue
  a = 2 *a
print(a)
```

## Nested Loops
When using `break` inside a nested loop, you can break only the inner one.

```python
while True:
  while True:
    break
  print("Only the inner loop is break.")
  break
```

## Exercise: Gretest Common Divisor (GCD) of Two Positive Integers

```python
num1,num2 = 922,592
```

Euclidean algorithm for calculating the GCD of two integers uses the fact that: $$\gcd(a,b)=\gcd(a-b,b)=\gcd(a,b-a)$$
and that:
$$\gcd(a,a)=a$$

```python
a, b = 15, 12
print(f"GCD({a},{b})",end=' = ')

while a != b:
  if a > b:
    print(f"GCD({a-b},{b})",end=' = ')
    a = a - b
  if a < b:
    print(f"GCD({a},{b-a})",end=' = ')
    b = b - a
else:
  print(a)
```

You can improve this algorithm by converting `if` statements to `while` statements.

```python
a, b = 15123, 1221
print(f"GCD({a},{b})",end=' = ')

while a != b:
  while a > b:
    a = a - b
  else:
    print(f"GCD({a},{b})",end=' = ')
  while a < b:
    b = b - a
  else:
    print(f"GCD({a},{b})",end=' = ')
else:
  print(a)
```

Also you can use the fact that
$$\gcd(a,b)=\gcd(a,b\% a)$$
and that:
$$\gcd(a,0)=a$$
Here you dont have to check if a < b or not in each step.
Instead just check if $b=0$ or not.

```python
a, b = 15123, 1221

print(f"GCD({a},{b}) is equal to the last item in the following sequence:")

while b:
  print(b, end=' ')
  a, b = b, a % b
```

## Exercise: Fibonacci Numbers

```python
print('Fibonacci numbers = ', end='')

a, b = 0, 1

while True:
  print(a, end=', ')
  a, b = b, a + b
  if a > 100:
    print('...')
    break
```

```python
print('Golden numbers = ', end='')

a, b = 1, 1

while True:
  print(f"{b}/{a}={b/a}", end=', ')
  a, b = b, a + b
  if a > 100:
    print('...')
    break
```
