### Mathematical Functions
To use mathematical functions you need to import `math` module once.

```python
import math
```

`math` module provides several mathematical functions.
To use them write `math` then a `.` and the name of the function.
Finally write the argument of that function in paranthesis.

Some of these functions are:

```python
# Square root of 16
math.sqrt(16)
```

```python
# Sine of 90 degrees (in radians)
math.sin(math.pi / 2)
```

```python
# Greatest Common Divisor of 48 and 180
math.gcd(48, 180)
```

```python
# Factorial of 5
math.factorial(5)
```

### Complex Functions
To use mathematical function on complex numbers import `cmath` module.

```python
import cmath
```

Some of the functions defined in `cmath` module are:

```python
# Get the square root of -16
cmath.sqrt(-16)
```

```python
# Get the cosine of the complex number 3 + 2j
cmath.cos(3 + 2j)
```

```python
# Get the polar coordinates (r, theta) of the complex number 3 + 4j
cmath.polar(3 + 4j)
```
