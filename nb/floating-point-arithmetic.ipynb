# Floating Point Arithmetic

A number is called _floating point number_ if it contains a `.`

```python
12.3
```

```python
.1
```

```python
12.
```

You can also use _scientific notation_ when writing floating point numbers:

```python
1.2354e3
```

Floating point arithmetic is very similar to integer arithmetic
but uses different algorithms for computation.
Usually floating point arithmetic is slower than integer arithmetic and performed in a dedicated circuit in the CPU.

When doing integer arithmetic
if any of the operands is a floating point number or the result can be a floating point number
then _floating point arithmetic_ is performed.

**Warning:** Floating point arithmetic sometimes gives surprising results.
For example:

```python
0.1 * 3
```

Therefore when working on floating point numbers you should know that the result may not be exact.

For (true) division use: `/`

```python
15.3 / 4.2
```

The (true) division is always done using floating point arithmetic.
Even if the result is integer.

```python
12 / 3
```

Unlike integers floating point numbers can **not** be arbitrary big.

```python
2.0**1023
```

Tu get some floating point number `number` rounded to `n` digits use `round(number,ndigits=n)`

```python
round(12.34567, ndigits=3)
```

## Complex Number Arithmetic
Python also supports complex number arithmetic.
Any complex number of the form $a+bi\in\mathbb C$ can be represented by `a+bj` in Python.

```python
((3 + 1j) * 2) ** 2
```

To get the real and imaginary part of a number use `.real` and `.imag` properties.

```python
(2 + 3j).real
```

```python
(2 + 3j).imag
```

To get the absolute value of a complex number use `abs()` function.

```python
abs(3+4j)
```
