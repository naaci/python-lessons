```python
print("enter two natural numbers")
a = int(input("a: "))
b = int(input("b: "))

if a < 0 or b < 0:
  raise ValueError("a,b must be non-negative")

while a != b:
  if a > b:
    a -= b
  if b> a:
    b -= a

print(a)
```

## Catching The Exceptions

```python
try:
  print(1/0)
except:
  print("something went wrong")
```

```python
try:
  a = []
  a[1]
  print(1/0)
except ZeroDivisionError as err:
  print("ZeroDivisionError:",err)
except IndexError as err:
  print("IndexError:",*err.args)
finally:
  print("ok")
```

```python
import logging
logging.basicConfig(
    filename="exceptions.log",
    level = logging.DEBUG,
    force=True
)

logging.getLogger().isEnabledFor(logging.DEBUG)
```

```python
try:
  a = []
  a[1]
except IndexError as err:
  logging.error(f"IndexError: {err}")
  logging.warning(f"IndexError: {err}")
  logging.info(f"IndexError: {err}")
  logging.debug(f"IndexError: {err}")
```
