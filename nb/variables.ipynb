# Variables
In Python variables are the names of the objects given by the programmer.

To give a name to a Python object just do an assignment with: `=`

```python
greeting = 'Hello, world!'
```

Now the string `'Hello, world!'` has the name `greeting`.
You can use `greeting` whenever you want to use that object.

```python
greeting
```

```python
greeting * 3
```

You can give multiple names to the same object.

```python
greeting_message = message = greeting
```

Now the object `Hello, world!` has three names: `greeting`, `message` and `greeting_message`.
All are the same object.
You can see this by asking with `is` keyword:

```python
greeting is greeting_message
```

**Warning:** here the keyword `is` does not do an equality check.
It says wheather two variables refer to the same object or not.

For example, if you write:

```python
greeting_message = 'Hello, world!'
```

Then `greeting is greeting_message` becomes `False`

```python
greeting is greeting_message
```

Because these are now refering to two different objects.

## Garbage Collection

Python objects remain in memory as soon as thay are refered by a variable.
If for some reason thay lost their references they are automatically removed from the memory by Python. This is called _garbage collection_.
You can delete a variable by `del` keyword.

```python
del greeting
```

The variable `greeting` is deleted but since the object referred by `greeting` still referred by another variable `message`, it is still in the memory.

```python
del message
```

Now it is deleted from memory.

## Naming Rules

In Python variable names contains letters, numbers and `_`.
But they can not start with a number.
For example `1a` can not be a variable name.

Also the following names can not be used as variable names:

`False`, `None`, `True`, `and`, `as`, `assert`, `async`, `await`, `break`, `class`, `continue`, `def`, `del`, `elif`, `else`, `except`, `finally`, `for`, `from`, `global`, `if`, `import`, `in`, `is`, `lambda`, `nonlocal`, `not`, `or`, `pass`, `raise`, `return`, `try`, `while`, `with`, `yield`

Because they have special meanings.

- Using meaninful names for variables help people to understand your code.
- Usually UPPERCASE variable names are used for constants.
- When the variable name has more than one word use `_` to seperate them.
